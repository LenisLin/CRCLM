## IMC analysis
follow the steps in *Xiao, X. et al. Multiplexed imaging mass cytometry reveals distinct tumor-immune microenvironments linked to immunotherapy responses in melanoma. Commun Med (Lond) 2, 131 (2022).* to construct analysis pipeline


```python
### 1. Convert raw data (.mcd) to TIFF (.tif) format
import imctools
from imctools.converters import mcdfolder2imcfolder
from imctools.converters import exportacquisitioncsv
from imctools.io.mcd.mcdparser import McdParser

import os
import shutil
import tifffile as tiff

BasePath='/mnt/data/lyx/IMC/'

folder_path_rawData=os.path.join(BasePath,'raw_data') # raw data
folder_path_MCD=os.path.join(BasePath,'mcd') # store every .mcd file in a single dir
folder_path_TIFF = os.path.join(BasePath,'tiffs')  #每个文件夹是一个ROI，里面是多个单通道信息
folder_path_ome= os.path.join(BasePath,'ometiff') #每个文件夹是一个sample，里面是多个ROI信息
folder_path_cp = os.path.join(BasePath,'cpout') #从cellprofiler出来的mask信息
folder_path_fullstack = os.path.join(BasePath,'fullstacks') #所有ROI信息
```

```python
## move all .mcd files to a dir 
if not os.path.exists(folder_path_MCD):
    os.mkdir(folder_path_MCD)
samples=[x for x in os.listdir(folder_path_rawData)]
for sample in samples:
    temPath1=os.path.join(folder_path_rawData,sample)
    files=os.listdir(temPath1)
    for file in files:
        if file.endswith(".mcd"):
            filename=file.split(".")[0]
            temPath2=os.path.join(folder_path_MCD,filename)
            if not os.path.exists(temPath2):
                os.mkdir(temPath2)
            shutil.copyfile(os.path.join(temPath1,file),os.path.join(temPath2,file))
            break;

del temPath1,temPath2,file,files,sample,samples,filename
```

```python
## .mcd to tiff
mcdfiles=os.listdir(folder_path_MCD)
if not os.path.exists(folder_path_TIFF):
    os.mkdir(folder_path_TIFF)

for mcdfile in mcdfiles:
    temPath1=os.path.join(folder_path_MCD,mcdfile)
    file=os.listdir(temPath1)[0]
    mcd = McdParser(os.path.join(temPath1,file))
    for mid in mcd.session.acquisition_ids:
        try:
            tem = mcd.get_acquisition_data(mid)
            tiff_folder = os.path.join(folder_path_TIFF,mcdfile+'_ROI'+str(mid))
            if not os.path.exists(tiff_folder):
                os.makedirs(tiff_folder)
            print('save '+tiff_folder)
            tem.save_tiffs(tiff_folder, compression=0, bigtiff=False)
        except:
                print(mcdfile, mid)
                continue
    mcd.close()
```

```python
## 删除测试能量通道
for roi in os.listdir(folder_path_TIFF):
    tiff_folder = os.path.join(folder_path_TIFF,roi)
    images = os.listdir(tiff_folder)
    if len(images) < 20: # energy file
        shutil.rmtree(tiff_folder)
del roi,file,mid,tem,temPath1,tiff_folder
```

```python
## convert .mcd to ometiff
mcdfiles=os.listdir(folder_path_MCD)
if not os.path.exists(folder_path_ome):
    os.mkdir(folder_path_ome)

for mcdfile in mcdfiles:
    temPath1=os.path.join(folder_path_MCD,mcdfile)
    print(f"The sample is {mcdfile}")
    mcdfolder2imcfolder.mcdfolder_to_imcfolder(temPath1, folder_path_ome, create_zip=False)
```

```python
## copy omeTIFF files into fullstack and rename
if not os.path.exists(folder_path_fullstack):
    os.mkdir(folder_path_fullstack)

omefiles=os.listdir(folder_path_ome)
for omefile in omefiles:
    temPath1=os.path.join(folder_path_ome,omefile)
    files=os.listdir(temPath1)
    files = [i for i in files if i.endswith(".tiff")]
    for file in files:
        sample = file.split('_s')[0]
        roi = file.split('_a')[1]

        temPath2=os.path.join(temPath1,file)
        img = tiff.imread(temPath2)
        fullstack_name = os.path.join(folder_path_fullstack, sample+'_ROI'+str(roi)+'_fullstacks.tiff')

        if (img.shape[0] > 3) and (not os.path.exists(fullstack_name)): # 不能量通道ROI
            print('save '+ fullstack_name)
            shutil.copyfile(temPath2,fullstack_name)
```

```python
## remove channel less than 5
files = os.listdir(folder_path_fullstack)
for file in files:
    img = tiff.imread(os.path.join(folder_path_fullstack,file))
    if (img.shape[0] < 5):
        os.remove(os.path.join(folder_path_fullstack,file))
```

```python
## generate panel table
import pandas as pd

### single roi
rois = os.listdir(folder_path_TIFF)
imgs = os.listdir(os.path.join(folder_path_TIFF, rois[1]))
imgs.sort()

panel = pd.DataFrame({'channel':[i.split('_')[2] for i in imgs]})

panel['marker']=[i[6:] for i in panel['channel']]
panelPath=os.path.join(BasePath,"IMC_CRC_panel.csv")
panel.to_csv(panelPath)
```
